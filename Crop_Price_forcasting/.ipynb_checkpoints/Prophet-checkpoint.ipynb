```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from numpy import percentile
from pandas.plotting import lag_plot
from statsmodels.tsa.stattools import adfuller
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from pandas import datetime
from bokeh.plotting import figure, output_notebook, show
```

# Generating data framespath

```python
# creating datframes for each crop and for each district
# paths of the data downloaded from agmarket.com (gov crops price website)
paths = ['Price_data_csvfiles/Rice.csv','Price_data_csvfiles/Maize.csv']
print('Data frame names:-')
# creating dataframes 
for i in paths:
    # downloaded data is in html format
    data = pd.read_csv(i)
#     data = data[data['Variety']=='Fine']
    data['date_time'] = pd.to_datetime(data['Price Date']) # converting into date time formate
    data['month_year'] = data['date_time'].dt.to_period('M')
    data.groupby('month_year').mean()
    data = data[['month_year','Max Price (Rs./Quintal)']] # keeping only required columns 
    data.rename(columns={'month_year':'date','Max Price (Rs./Quintal)':'price'},inplace=True) # renaming columns names
    ## Outliers
    q25, q75 = percentile(data['price'], 25), percentile(data['price'], 75)
    iqr = q75 - q25
    # calculate the outlier cutoff
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off
    # identify outliers
    outliers = [x for x in data['price'] if x < lower or x > upper]
    outliers = list(dict.fromkeys(outliers))
    Removing_outliers = [data.drop(data[data['price']==x].index,axis=0,inplace=True) for x in outliers]
    if len(Removing_outliers)==0:print(f'No outliers in the  dataframe')
    else:print(f'{len(Removing_outliers)} outliers has been detected in  dataframe and removed')
    data = data.groupby('date').mean()# grouping into month level
    data['price'] = data['price'].round()
    data = data.reindex(pd.period_range(data.index[0],data.index[-1],freq='M'))# adding the missing rows into the data
    data = data.reset_index()
    data['index']= data['index'].astype(str)
    data['index'] = pd.to_datetime(data['index'])
    data = data.set_index('index')
    globals()[i.split('/')[1].split('.')[0]] = data 
    print('  ',i.split('/')[1].split('.')[0])
```

```python
Maize['price'] = Maize['price'].fillna(method='ffill')
Rice['price'] = Rice['price'].fillna(method='ffill')
```
